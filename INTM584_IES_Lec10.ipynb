```python
import pandas as pd
df=pd.read_csv("/content/TCS.NS.csv")
df
```

```python
#linear Regression
x=df.iloc[:,0:1].values #entire data divided into small small arrays, without 0:1 single array
x
```

```python
y=df.iloc[:,4].values
y
```

```python
from sklearn.linear_model import LinearRegression
Regressor=LinearRegression()
Regressor.fit(x,y)
```

```python
y_pred = Regressor.predict(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.scatter(x,y,color="Red")
```

```python
plt.scatter(x,y,color="Red")
plt.plot(x,y_pred) #Regression lines
```

```python
import seaborn as sns
sns.lmplot(x="Date", y="Close",data=df)
sns.lineplot(x="Date", y=y_pred , data=df)
sns.set_style("dark")
#additional line generated from autocorrelation, helps us to find accuracy of max. changed range, the variance
```

```python
import seaborn as sns
sns.swarmplot(x="Date", y="Close",data=df)
sns.lineplot(x="Date", y=y_pred , data=df)
sns.set_style("dark")
```
